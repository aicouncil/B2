```python
df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
d1 = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
d1.insert(0, 'const' , 1)
d1
```

```python
vif_data = pd.DataFrame()
vif_data['features'] = d1.columns

vif_data
```

```python
vif_values = []
for i in range(5):
  #print(variance_inflation_factor(d1, i))
  vif_values.append(variance_inflation_factor(d1, i))
```

```python
vif_data['VIF'] = vif_values
```

```python
vif_data
```

**VIF score**

  * vif <= 1 -> feature is poorly corealted with other features, that feature should not be dropped
  * vif <= 5 -> we can use by monitoring the model performance
  * vif > 5 -> we generally drop the feature as others features are highly corealted with this feature and they know about it very well

```python
d2 = d1.drop(columns = 'PetalLengthCm')
d2
```

```python
vif_data = pd.DataFrame()
vif_data['features'] = d2.columns

vif_values = []
for i in range(len(d2.columns)):
  #print(variance_inflation_factor(d1, i))
  vif_values.append(variance_inflation_factor(d2, i))

vif_data['VIF'] = vif_values
```

```python
X = df[['SepalLengthCm', 	'SepalWidthCm', 'PetalWidthCm']]
y = df['Species']
```
