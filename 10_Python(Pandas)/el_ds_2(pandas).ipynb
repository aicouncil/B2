**Pandas**

  * Pandas is a powerful python library used for data manipulation and analysis.
  * Pandas - Panel Data
  * Built on top of Numpy
  * Often used for data science, machine learning , finance , healtcare.

**Comapare Numpy and Pandas**

  * Purpose
    * Numpy - Numerical computing
    * Pandas - Data analysis and manipulation
  * Data structure
    * Numpy - ndarray (multi-dimensional array)
    * Series (1D) & Dataframe (2D like excel)
  * Data type
    * Numpy - Homogeneous
    * Pandas - Hetrogeneous (mixed type)
  * Speed
    * Numpy - faster for numeric computation
    * Pandas - Slightlt slower but more powerful for tables

**Identify the region of 1st , 2nd and 3rd standard deviation of Age**

```python
df['Age'].hist(bins=20)
```

```python
df['Age'].mean()
```

```python
(df['Age'] < 29.70).sum()/len(df)
```

```python
(df['Age'] > 29.70).sum()/len(df)
```

```python
mean_age = df['Age'].mean()

std_age = df['Age'].std()
```

```python
#1st standard deviation
lower_limit_1st_std = mean_age - std_age
upper_limit_1st_std = mean_age + std_age
print(lower_limit_1st_std , upper_limit_1st_std)
```

```python
#%ge of passengers in 1st std deviation
((df['Age'] > lower_limit_1st_std) & (df['Age'] < upper_limit_1st_std)).sum()/len(df)
```

```python
#2nd standard deviation
lower_limit_2nd_std = mean_age - 2 * std_age
upper_limit_2nd_std = mean_age + 2 * std_age
print(lower_limit_2nd_std , upper_limit_2nd_std)
```

```python
#%ge of passengers in 2nd standard deviation
((df['Age'] > lower_limit_2nd_std) & (df['Age'] < upper_limit_2nd_std)).sum()/len(df)
```

```python
#3rd standard deviation
lower_limit_3rd_std = mean_age - 3 * std_age
upper_limit_3rd_std = mean_age + 3 * std_age
print(lower_limit_3rd_std , upper_limit_3rd_std)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'])

plt.axvline(x=mean_age, c='black')

plt.axvline(x=lower_limit_1st_std , c='g')
plt.axvline(x=upper_limit_1st_std , c='g')

plt.axvline(x=lower_limit_2nd_std , c='y')
plt.axvline(x=upper_limit_2nd_std , c='y')

plt.axvline(x=lower_limit_3rd_std , c='r')
plt.axvline(x=upper_limit_3rd_std , c='r')

plt.show()
```
