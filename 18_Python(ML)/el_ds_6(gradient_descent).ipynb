**Error optimization process in machine learning**

  * Gradient descent


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/refs/heads/main/Salary_dataset.csv',
                 index_col=0)
df.head(10)
```

```python
X = df['YearsExperience']
y = df['Salary']
```

```python
w0 = 1
w1 = 1
yh = w0 + w1 * X
```

```python
error = ((y - yh)**2).mean()
print(error)
```

```python
dew0 = -2*((y - yh)).mean()
print(dew0)

print()

dew1 = -2*((y - yh)*X).mean()
print(dew1)
```

```python
lr = 1e-6

w0 = w0 - lr * dew0
w1 = w1 - lr * dew1
```

```python
print(w0 , w1)
```

```python
yh = w0 + w1 * X
error = ((y - yh)**2).mean()
print(error)
```

```python
w0 = 1
w1 = 1

for i in range(5000):
  yh = w0 + w1 * X

  dew0 = -2*((y - yh)).mean()
  dew1 = -2*((y - yh)*X).mean()

  lr = 1e-2

  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1

  error = ((y - yh)**2).mean()
  print(i,error)
```

```python
print(w0 , w1)
```

```python
#Yh = w0 + w1 * X
yoe = 5.3

predicted_salary = w0 + w1 * yoe
print(predicted_salary)
```

```python
print("Mean absolute error")
print(abs(y - yh).mean())
```

```python
print("R2 score")
print(1 - (((y - yh)**2).sum()/((y - y.mean())**2).sum()))
```
