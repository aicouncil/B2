**Pandas**

  * Pandas is a powerful python library used for data manipulation and analysis.
  * Pandas - Panel Data
  * Built on top of Numpy
  * Often used for data science, machine learning , finance , healtcare.

**Comapare Numpy and Pandas**

  * Purpose
    * Numpy - Numerical computing
    * Pandas - Data analysis and manipulation
  * Data structure
    * Numpy - ndarray (multi-dimensional array)
    * Series (1D) & Dataframe (2D like excel)
  * Data type
    * Numpy - Homogeneous
    * Pandas - Hetrogeneous (mixed type)
  * Speed
    * Numpy - faster for numeric computation
    * Pandas - Slightlt slower but more powerful for tables

**Series - 1D**

```python
# Series - 1D
s = pd.Series([10,20,30,40])
s.columns = ['a']
s.index = ['a','b','c','d']      #Label based indexing
s
```

* Series is 1- dimensional and doesn't have columns
* Series has only index and values
* So, trying to assign .columns => it will raise an error or be ignored silently

```python
print(type(s))
```

```python
print(s.loc['a'])
```

```python
#data filter where age is more then 30
df_age_more_30 = df[df['Age'] > 30]
df_age_more_30.head()
```

```python
#filter Survived male passengers whose age is more then 30

df_male_survived_more_then_30 = df[(df['Survived'] == 1) & (df['Sex'] == 'male') & (df['Age'] > 30)]
df_male_survived_more_then_30
```
