**Non-Linear Model for advertising data**

```python
df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/refs/heads/main/Advertising.csv',
                 index_col=0)
df
```

**Non-linear model using TV as feature**

```python
x = df[['TV']]
y = df['sales']
```

```python
#prepare data for 2-degree non linear regression

x_nl = pd.concat((x, x**2, x**3),axis=1)
x_nl.columns = ['TV', 'TV_squared', 'TV_cube']
x_nl
```

```python
from sklearn.linear_model import LinearRegression

model_nl_tv = LinearRegression()

model_nl_tv.fit(x_nl,y)
```

```python
model_nl_tv.coef_
```

```python
#Evaluate model performance

ypred_nl = model_nl_tv.predict(x_nl)
```

```python
mae_nl_model = abs(y - ypred_nl).mean()
print(mae_nl_model )
```

**Apply 2 degree non-linear transformation on advertising data (TV,radio) and build a predictive model**

  * compare the non-linear model performance with linear model performance that we created earlier

```python
X = df[['TV','radio']]
y = df['sales']
```

**To test Overfitting**

    * Split your data into two parts
        - data for training - (75%)
        - data for testing - (25%)

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.75)
```

```python
print(X.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol = PolynomialFeatures(degree=2 , include_bias=False)

pol.fit(xtrain)        #let the PolynomialFeatures know, we have 2 features to be transformed into 5 features fit for 2 degree non-linear equation

Xtrain_pol = pol.transform(xtrain)
xtest_pol = pol.transform(xtest)
```

```python
print(Xtrain_pol.shape)
print(xtest_pol.shape)
```

```python
from sklearn.linear_model import LinearRegression
model_nl_adv = LinearRegression()
model_nl_adv.fit(Xtrain_pol, ytrain)
```

```python
print(model_nl_adv.coef_)
print(model_nl_adv.intercept_)
```

```python
ypredTrain_nl = model_nl_adv.predict(Xtrain_pol)      #prediction on training data => data used during training process

ypredTest_nl = model_nl_adv.predict(xtest_pol)          #prediction on test data => data not used for training process
```

```python
#Mean absolute error on training data
mae_Train = abs(ytrain - ypredTrain_nl).mean()
print(mae_Train)
```

```python
#Mean absolute error on test data
mae_test = abs(ytest - ypredTest_nl).mean()
print(mae_test)
```

```python
tv_expense = 30.5
radio_expense = 70.9

print(model_nl_adv.predict([[tv_expense , radio_expense , tv_expense**2 , tv_expense*radio_expense , radio_expense**2]]))
```

**Conclusion**

  * As train and test error values are very close so , we build a non-overfit model

**Can we improve model performance without increasing the size of data much**

  * Yes, we can do it using feature interaction.

```python
X = df[['TV','radio']]
y = df['sales']
```

```python
X_new = X.copy()
X_new['TVR'] = X.TV * X.radio
```

```python
X_new.head()
```

```python
#Split the data into train & test
from sklearn.model_selection import train_test_split
xtrain_new,xtest_new,ytrain_new,ytest_new = train_test_split(X_new,y,train_size=0.75)
```

```python
print(X_new.shape)
print(xtrain_new.shape)
print(xtest_new.shape)
```

```python
from sklearn.linear_model import LinearRegression
model_fi = LinearRegression()

model_fi.fit(xtrain_new,ytrain_new)
```

```python
print(model_fi.coef_)
```

```python
#prediction made on training data
ytrainPred = model_fi.predict(xtrain_new)

#prediction on test data
ytestPred = model_fi.predict(xtest_new)
```

```python
maeTrain_fi = abs(ytrain_new - ytrainPred).mean()
print(maeTrain_fi)
```

```python
maeTest_fi = abs(ytest_new - ytestPred).mean()
print(maeTest_fi)
```

```python
tv_expense = 30.5
radio_expense = 70.9

print(model_fi.predict([[tv_expense , radio_expense , tv_expense*radio_expense]]))
```
